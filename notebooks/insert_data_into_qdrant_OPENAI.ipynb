```python

# Qdrant server URL
URL ="localhost"
# Qdrant dimension of the collection
DIMENSION = 1536
# Qdrant collection name
COLLECTION_NAME = "STARTUPS_OPENAI_EMBEDDINGS"
METRIC_NAME ="COSINE"
```

```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import qdrant_client as qc
import qdrant_client.http.models as qmodels
from qdrant_client.http.models import *
from openai_embeddings import get_embeddings_azure_openai

import json
import time
```

```python
df = pd.read_json('./startups_demo.json', lines=True)
```

```python
Lines = []
payload_list = []
counter =0
```

```python
for row in df.itertuples():
    Lines.append(row.alt + "." + row.description)
    payload_item ={
        "name": row.name,
        "images": row.images,
        "alt": row.alt,
        "description": row.description,
        "city": row.city,
        "link": row.link,
            }
    payload_list.append(payload_item)
    counter += 1
    if counter % 500 == 0:
        break;
    
```

```python
len(Lines)
```

```python
Lines[0]
```

```python
embeddings_all = []
```

```python
%%time
counter = 0 
for line in Lines:
    embedding = get_embeddings_azure_openai(line)
    embeddings_all.append(embedding)
    counter += 1
    if counter % 100 == 0:
        time.sleep(10)
```

```python
len(embeddings_all)
```

```python
len(embeddings_all[9])
```

```python
qdrant_client = qc.QdrantClient(URL)
```

```python
qdrant_client.upload_collection(
    collection_name=COLLECTION_NAME,
    vectors=embeddings_all,
    payload=payload_list,
    ids=None,  # Vector ids will be assigned automatically
    batch_size=256  # How many vectors will be uploaded in a single request?
)
```
